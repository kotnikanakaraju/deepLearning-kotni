```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
mark=['Sometimes to understand a word',
      'meaning you need more than a definition;',
      'you need to see the word used in a sentence.',
      'At YourDictionary,',
      'we give you the tools to learn what a word means',
      'and how to use it correctly.',
      ' With this sentence maker,'
      'simply type a word in the search bar and see a variety of sentences',
      'with that word used in its different ways.',
      'Our sentence generator can provide more context and relevance,',
      'ensuring you use a word the right',]
```

```python
# size of mark
mark
voc_size=9000
```

# one hot representation

```python
onehot_rep=[one_hot(words,voc_size) for words in mark]
print(onehot_rep)
```

# word embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
mark_length=10
docs=pad_sequences(onehot_rep,padding='pre',maxlen=mark_length)
print(docs)
```

```python
dim=12
model=Sequential()
model.add(Embedding(voc_size,12,input_length=mark_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(docs))
```
